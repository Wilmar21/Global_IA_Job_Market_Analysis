```python
# Interpretacion de las distribuciones de las variables numericas
import matplotlib.pyplot as plt
import seaborn as sns
import math
import scipy.stats as stats
import numpy as np

# Histogramas
jobsDatesetColumnasNumericas = jobsDatesetOriginal.drop(columns=['job_id', 'job_title', 'salary_currency', 'experience_level', 'employment_type', 'company_location', 'company_size', 'employee_residence', 'required_skills','education_required','industry','posting_date','application_deadline','company_name'])

plt.figure(figsize=(12, 10))
jobsDatesetColumnasNumericas.hist(bins=30, figsize=(12, 10))

plt.suptitle('Distribucion de las variables numericas IA JOBS - Histogramas')
plt.tight_layout()
plt.show()

# QQ Plots
numeric_columns = jobsDatesetColumnasNumericas.select_dtypes(include='number').columns
num_cols = len(numeric_columns)
cols = 2
rows = math.ceil(num_cols / cols)
plt.figure(figsize=(cols * 6, rows * 4))

for i, column in enumerate(numeric_columns, 1):
    plt.subplot(rows, cols, i)
    stats.probplot(jobsDatesetColumnasNumericas[column], dist="norm", plot=plt)
    plt.title(f"QQ Plot - {column}")

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.suptitle('QQ Plots de Variables Numéricas', fontsize=16)
plt.show()

# Subplots
numeric_columns = jobsDatesetColumnasNumericas.select_dtypes(include='number').columns
num_cols = len(numeric_columns)
cols = 2
rows = math.ceil(num_cols / cols)
plt.figure(figsize=(12, rows * 4))

for i, column in enumerate(numeric_columns, 1):
    plt.subplot(rows, cols, i)
    sns.boxplot(data=jobsDatesetColumnasNumericas[column], color='skyblue')
    plt.title(column)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.suptitle("Distribución de las Variables Numéricas en el Dataset de IA JOBS", fontsize=16)
plt.show()
```

```python
# Interpretacion de las distribuciones de las variables Categoricas
jobsDatesetColumnasCategoricas = jobsDatesetOriginal.drop(columns=['salary_usd', 'remote_ratio', 'years_experience', 'job_description_length', 'benefits_score'])

jobsDatesetColumnasCategoricas.info()

# nivel de experiencia
plt.figure(figsize=(10, 4))
sns.countplot(y='experience_level', data=jobsDatesetColumnasCategoricas, order=jobsDatesetColumnasCategoricas['experience_level'].value_counts().index, hue='experience_level', dodge=False)
plt.title('Distribución IA JOBS', fontsize=16)
plt.xlabel('Cantidad de IA JOBS', fontsize=14)
plt.ylabel('experience_level', fontsize=14)
plt.legend([], [], frameon=False)  # Oculta la leyenda innecesaria
plt.show()

# nivel de educacion requerido
plt.figure(figsize=(10, 4))
sns.countplot(y='education_required', data=jobsDatesetColumnasCategoricas, order=jobsDatesetColumnasCategoricas['education_required'].value_counts().index, hue='education_required', dodge=False)
plt.title('Distribución IA JOBS', fontsize=16)
plt.xlabel('Cantidad de IA JOBS', fontsize=14)
plt.ylabel('education_required', fontsize=14)
plt.legend([], [], frameon=False)  # Oculta la leyenda innecesaria
plt.show()

# tipo de empleo
plt.figure(figsize=(10, 4))
sns.countplot(y='employment_type', data=jobsDatesetColumnasCategoricas, order=jobsDatesetColumnasCategoricas['employment_type'].value_counts().index, hue='employment_type', dodge=False)
plt.title('Distribución IA JOBS', fontsize=16)
plt.xlabel('Cantidad de IA JOBS', fontsize=14)
plt.ylabel('employment_type', fontsize=14)
plt.legend([], [], frameon=False)  # Oculta la leyenda innecesaria
plt.show()

# nombre de la empresa
plt.figure(figsize=(10, 4))
sns.countplot(y='company_name', data=jobsDatesetColumnasCategoricas, order=jobsDatesetColumnasCategoricas['company_name'].value_counts().index, hue='company_name', dodge=False)
plt.title('Distribución IA JOBS', fontsize=16)
plt.xlabel('Cantidad de IA JOBS', fontsize=14)
plt.ylabel('company_name', fontsize=14)
plt.legend([], [], frameon=False)  # Oculta la leyenda innecesaria
plt.show()

# Comparando el numero combinaciones entre 2 columnas
contingency_table = pandas.crosstab(jobsDatesetColumnasCategoricas['education_required'], jobsDatesetColumnasCategoricas['experience_level'])
plt.figure(figsize=(14, 8))
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='viridis')
plt.title('Relación entre la educacion requerida y el nivel de experiencia')
plt.xlabel('Nivel de experiencia')
plt.ylabel('Educacion requerida')
plt.show()

contingency_table = pandas.crosstab(jobsDatesetColumnasCategoricas['employment_type'], jobsDatesetColumnasCategoricas['experience_level'])
plt.figure(figsize=(14, 8))
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='viridis')
plt.title('Relación entre el tipo de empleo y el nivel de experiencia')
plt.xlabel('Nivel de experiencia')
plt.ylabel('Tipo de empleo')
plt.show()

contingency_table = pandas.crosstab(jobsDatesetColumnasCategoricas['education_required'], jobsDatesetColumnasCategoricas['company_size'])
plt.figure(figsize=(14, 8))
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='viridis')
plt.title('Relación entre la educacion requerida y el tamaño de la empresa')
plt.xlabel('Tamaño de la empresa')
plt.ylabel('Educacion requerida')
plt.show()

contingency_table = pandas.crosstab(jobsDatesetOriginal['employment_type'], jobsDatesetOriginal['remote_ratio'])
plt.figure(figsize=(14, 8))
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='viridis')
plt.title('Relación entre el tipo de empleo y el porcentaje de remoto')
plt.xlabel('Porcentaje de remoto')
plt.ylabel('Tipo de empleo')
plt.show()
```

```python
# Convertir Columnas categoricas a columnas numericas para poder interpretar su correlación
jobsDatasetClean['job_title'] = encoder.fit_transform(jobsDatasetClean['job_title'])
jobsDatasetClean['employment_type'] = encoder.fit_transform(jobsDatasetClean['employment_type'])
jobsDatasetClean['company_location'] = encoder.fit_transform(jobsDatasetClean['company_location'])
jobsDatasetClean['industry'] = encoder.fit_transform(jobsDatasetClean['industry'])
jobsDatasetClean['education_required'] = encoder.fit_transform(jobsDatasetClean['education_required'])
jobsDatasetClean['experience_level'] = encoder.fit_transform(jobsDatasetClean['experience_level'])
jobsDatasetClean['company_size'] = encoder.fit_transform(jobsDatasetClean['company_size'])

# Calcular la matriz de correlación
numeric_df = jobsDatasetClean.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_df.corr()
plt.figure(figsize=(24, 16))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Matriz de Correlación', fontsize=16)
plt.show()
```

```python
# 5. Detección y manejo de valores atípicos
from scipy.stats import zscore
import numpy as np
from pyod.models.mad import MAD

jobsDatasetCleanOutliers = jobsDatasetClean.copy()

# Histogramas
plt.figure(figsize=(12, 10))
jobsDatasetCleanOutliers.hist(bins=30, figsize=(12, 10))
plt.suptitle('Distribucion de las variables numericas IA JOBS - Histogramas')
plt.tight_layout()
plt.show()
# boxplots
numeric_columns = jobsDatasetCleanOutliers.select_dtypes(include='number').columns
num_cols = len(numeric_columns)
cols = 2
rows = math.ceil(num_cols / cols)
plt.figure(figsize=(12, rows * 4))

for i, column in enumerate(numeric_columns, 1):
    plt.subplot(rows, cols, i)
    sns.boxplot(data=jobsDatasetCleanOutliers[column], color='skyblue')
    plt.title(column)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.suptitle("Distribución de las Variables Numéricas en el Dataset de IA JOBS", fontsize=16)
plt.show()

print('VALORES ORIGINALES PARA LA VARIABLE SALARIO')
columna = jobsDatasetCleanOutliers['salary_usd']
plt.figure(figsize=(6, 4))  # Tamaño de la figura
columna.hist(bins=30)  # Puedes cambiar el número de bins según lo que necesites
plt.title('Histograma Amount')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
sns.boxplot(x=jobsDatasetCleanOutliers['salary_usd'])
plt.show()

# Aplicacion modelo Z-SCORE para el manejo de outliers
print('VALORES Z-Score')
jobsDatasetCleanOutliers['zscore'] = zscore(jobsDatasetCleanOutliers['salary_usd'])
outliers = jobsDatasetCleanOutliers[jobsDatasetCleanOutliers['zscore'].abs() > 3]
z_scores = zscore(jobsDatasetCleanOutliers['salary_usd'])
abs_z_scores = np.abs(z_scores)
print(f'Numero de outliers Z-Score: {len(outliers)}')

jobsDatasetConZScore = pandas.merge(jobsDatasetCleanOutliers, outliers, how='left', indicator=True).query('_merge == "left_only"').drop('_merge', axis=1)

columna = jobsDatasetConZScore['salary_usd']
plt.figure(figsize=(6, 4))  # Tamaño de la figura
columna.hist(bins=30)  # Puedes cambiar el número de bins según lo que necesites
plt.title('Histograma Amount diferencia_1')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
sns.boxplot(x=jobsDatasetConZScore['salary_usd'])
plt.show()

# Aplicacion modelo Z-SCORE mejorado para el manejo de outliers
print('VALORES Z-Score Mejorado')
mad = MAD(threshold = 3.5)
total_reshaped = jobsDatasetCleanOutliers['salary_usd'].values.reshape(-1, 1)
labels = mad.fit(total_reshaped).labels_
print(f'Numero de outliers Z-Score Modificado: {labels.sum()}')

jobsDatasetConZScoreMejorado = jobsDatasetCleanOutliers.copy()
jobsDatasetConZScoreMejorado = jobsDatasetConZScoreMejorado[labels == 0]
columna = jobsDatasetConZScoreMejorado['salary_usd']
plt.figure(figsize=(6, 4))
columna.hist(bins=30)
plt.title('Histograma Amount')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()

sns.boxplot(x=jobsDatasetConZScoreMejorado['salary_usd'])
plt.show()

# Aplicacion modelo IQR - Rango intercuartilico para el manejo de outliers
print('VALORES IQR - Rango intercuartílico')
Q1 = jobsDatasetCleanOutliers['salary_usd'].quantile(0.25)
Q3 = jobsDatasetCleanOutliers['salary_usd'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

jobsDatasetConIQR = jobsDatasetCleanOutliers[(jobsDatasetCleanOutliers['salary_usd'] < lower_bound) | (jobsDatasetCleanOutliers['salary_usd'] > upper_bound)]
print("Outliers según IQR:")
print(len(jobsDatasetConIQR))

DF_jobsDatasetConIQR = jobsDatasetCleanOutliers[(jobsDatasetCleanOutliers['salary_usd'] >= lower_bound) & (jobsDatasetCleanOutliers['salary_usd'] <= upper_bound)]
print("Datos sin outliers:")
print(len(DF_jobsDatasetConIQR))

columna = DF_jobsDatasetConIQR['salary_usd']
plt.figure(figsize=(6, 4))  # Tamaño de la figura
columna.hist(bins=30)  # Puedes cambiar el número de bins según lo que necesites
plt.title('Histograma Amount')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()

sns.boxplot(x=DF_jobsDatasetConIQR['salary_usd'])
plt.show()
```
