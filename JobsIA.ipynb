```python
import pandas
# ANALISIS EXPLORATORIO EDA - DATASET IA_JOBS
import pandas as pd
jobsDatesetOriginal = pandas.read_csv('ai_job_dataset.csv', encoding='utf-8', low_memory=False)

# Mostramos los primeros datos del dataset
print("DATASET IA JOBS")
jobsDatesetOriginal.head()
```

```python
# Interpretacion de las distribuciones de las variables numericas
import matplotlib.pyplot as plt
import seaborn as sns
import math
import scipy.stats as stats
import numpy as np

# Histogramas
jobsDatesetColumnasNumericas = jobsDatesetOriginal.drop(columns=['job_id', 'job_title', 'salary_currency', 'experience_level', 'employment_type', 'company_location', 'company_size', 'employee_residence', 'required_skills','education_required','industry','posting_date','application_deadline','company_name'])

plt.figure(figsize=(12, 10))
jobsDatesetColumnasNumericas.hist(bins=30, figsize=(12, 10))

plt.suptitle('Distribucion de las variables numericas IA JOBS - Histogramas')
plt.tight_layout()
plt.show()

# QQ Plots
numeric_columns = jobsDatesetColumnasNumericas.select_dtypes(include='number').columns
num_cols = len(numeric_columns)
cols = 2
rows = math.ceil(num_cols / cols)
plt.figure(figsize=(cols * 6, rows * 4))

for i, column in enumerate(numeric_columns, 1):
    plt.subplot(rows, cols, i)
    stats.probplot(jobsDatesetColumnasNumericas[column], dist="norm", plot=plt)
    plt.title(f"QQ Plot - {column}")

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.suptitle('QQ Plots de Variables Numéricas', fontsize=16)
plt.show()

# Subplots
numeric_columns = jobsDatesetColumnasNumericas.select_dtypes(include='number').columns
num_cols = len(numeric_columns)
cols = 2
rows = math.ceil(num_cols / cols)
plt.figure(figsize=(12, rows * 4))

for i, column in enumerate(numeric_columns, 1):
    plt.subplot(rows, cols, i)
    sns.boxplot(data=jobsDatesetColumnasNumericas[column], color='skyblue')
    plt.title(column)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.suptitle("Distribución de las Variables Numéricas en el Dataset de Diabetes", fontsize=16)
plt.show()
```

```python
# Interpretacion de las distribuciones de las variables Categoricas
jobsDatesetColumnasCategoricas = jobsDatesetOriginal.drop(columns=['salary_usd', 'remote_ratio', 'years_experience', 'job_description_length', 'benefits_score'])

jobsDatesetColumnasCategoricas.info()

# nivel de experiencia
plt.figure(figsize=(10, 4))
sns.countplot(y='experience_level', data=jobsDatesetColumnasCategoricas, order=jobsDatesetColumnasCategoricas['experience_level'].value_counts().index, hue='experience_level', dodge=False)
plt.title('Distribución IA JOBS', fontsize=16)
plt.xlabel('Cantidad de IA JOBS', fontsize=14)
plt.ylabel('experience_level', fontsize=14)
plt.legend([], [], frameon=False)  # Oculta la leyenda innecesaria
plt.show()

# nivel de educacion requerido
plt.figure(figsize=(10, 4))
sns.countplot(y='education_required', data=jobsDatesetColumnasCategoricas, order=jobsDatesetColumnasCategoricas['education_required'].value_counts().index, hue='education_required', dodge=False)
plt.title('Distribución IA JOBS', fontsize=16)
plt.xlabel('Cantidad de IA JOBS', fontsize=14)
plt.ylabel('education_required', fontsize=14)
plt.legend([], [], frameon=False)  # Oculta la leyenda innecesaria
plt.show()

# tipo de empleo
plt.figure(figsize=(10, 4))
sns.countplot(y='employment_type', data=jobsDatesetColumnasCategoricas, order=jobsDatesetColumnasCategoricas['employment_type'].value_counts().index, hue='employment_type', dodge=False)
plt.title('Distribución IA JOBS', fontsize=16)
plt.xlabel('Cantidad de IA JOBS', fontsize=14)
plt.ylabel('employment_type', fontsize=14)
plt.legend([], [], frameon=False)  # Oculta la leyenda innecesaria
plt.show()

# Comparando el numero combinaciones entre 2 columnas
contingency_table = pandas.crosstab(jobsDatesetColumnasCategoricas['education_required'], jobsDatesetColumnasCategoricas['experience_level'])
plt.figure(figsize=(14, 8))
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='viridis')
plt.title('Relación entre la educacion requerida y el nivel de experiencia')
plt.xlabel('Nivel de experiencia')
plt.ylabel('Educacion requerida')
plt.show()

contingency_table = pandas.crosstab(jobsDatesetColumnasCategoricas['employment_type'], jobsDatesetColumnasCategoricas['experience_level'])
plt.figure(figsize=(14, 8))
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='viridis')
plt.title('Relación entre el tipo de empleo y el nivel de experiencia')
plt.xlabel('Nivel de experiencia')
plt.ylabel('Tipo de empleo')
plt.show()

contingency_table = pandas.crosstab(jobsDatesetColumnasCategoricas['education_required'], jobsDatesetColumnasCategoricas['company_size'])
plt.figure(figsize=(14, 8))
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='viridis')
plt.title('Relación entre la educacion requerida y el tamaño de la empresa')
plt.xlabel('Tamaño de la empresa')
plt.ylabel('Educacion requerida')
plt.show()

contingency_table = pandas.crosstab(jobsDatesetOriginal['employment_type'], jobsDatesetOriginal['remote_ratio'])
plt.figure(figsize=(14, 8))
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='viridis')
plt.title('Relación entre el tipo de empleo y el porcentaje de remoto')
plt.xlabel('Porcentaje de remoto')
plt.ylabel('Tipo de empleo')
plt.show()
```
