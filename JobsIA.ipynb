```python
# 5. Detección y manejo de valores atípicos
from scipy.stats import zscore
import numpy as np
from pyod.models.mad import MAD

jobsDatasetCleanOutliers = jobsDatasetClean.copy()

# Histogramas
plt.figure(figsize=(12, 10))
jobsDatasetCleanOutliers.hist(bins=30, figsize=(12, 10))
plt.suptitle('Distribucion de las variables numericas IA JOBS - Histogramas')
plt.tight_layout()
plt.show()
# boxplots
numeric_columns = jobsDatasetCleanOutliers.select_dtypes(include='number').columns
num_cols = len(numeric_columns)
cols = 2
rows = math.ceil(num_cols / cols)
plt.figure(figsize=(12, rows * 4))

for i, column in enumerate(numeric_columns, 1):
    plt.subplot(rows, cols, i)
    sns.boxplot(data=jobsDatasetCleanOutliers[column], color='skyblue')
    plt.title(column)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.suptitle("Distribución de las Variables Numéricas en el Dataset de IA JOBS", fontsize=16)
plt.show()

print('VALORES ORIGINALES PARA LA VARIABLE SALARIO')
columna = jobsDatasetCleanOutliers['salary_usd']
plt.figure(figsize=(6, 4))  # Tamaño de la figura
columna.hist(bins=30)  # Puedes cambiar el número de bins según lo que necesites
plt.title('Histograma Amount')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
sns.boxplot(x=jobsDatasetCleanOutliers['salary_usd'])
plt.show()

# Aplicacion modelo Z-SCORE para el manejo de outliers
print('VALORES Z-Score')
jobsDatasetCleanOutliers['zscore'] = zscore(jobsDatasetCleanOutliers['salary_usd'])
outliers = jobsDatasetCleanOutliers[jobsDatasetCleanOutliers['zscore'].abs() > 3]
z_scores = zscore(jobsDatasetCleanOutliers['salary_usd'])
abs_z_scores = np.abs(z_scores)
print(f'Numero de outliers Z-Score: {len(outliers)}')

jobsDatasetConZScore = pandas.merge(jobsDatasetCleanOutliers, outliers, how='left', indicator=True).query('_merge == "left_only"').drop('_merge', axis=1)

columna = jobsDatasetConZScore['salary_usd']
plt.figure(figsize=(6, 4))  # Tamaño de la figura
columna.hist(bins=30)  # Puedes cambiar el número de bins según lo que necesites
plt.title('Histograma Amount diferencia_1')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
sns.boxplot(x=jobsDatasetConZScore['salary_usd'])
plt.show()

# Aplicacion modelo Z-SCORE mejorado para el manejo de outliers
print('VALORES Z-Score Mejorado')
mad = MAD(threshold = 3.5)
total_reshaped = jobsDatasetCleanOutliers['salary_usd'].values.reshape(-1, 1)
labels = mad.fit(total_reshaped).labels_
print(f'Numero de outliers Z-Score Modificado: {labels.sum()}')

jobsDatasetConZScoreMejorado = jobsDatasetCleanOutliers.copy()
jobsDatasetConZScoreMejorado = jobsDatasetConZScoreMejorado[labels == 0]
columna = jobsDatasetConZScoreMejorado['salary_usd']
plt.figure(figsize=(6, 4))
columna.hist(bins=30)
plt.title('Histograma Amount')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()

sns.boxplot(x=jobsDatasetConZScoreMejorado['salary_usd'])
plt.show()

# Aplicacion modelo IQR - Rango intercuartilico para el manejo de outliers
print('VALORES IQR - Rango intercuartílico')
Q1 = jobsDatasetCleanOutliers['salary_usd'].quantile(0.25)
Q3 = jobsDatasetCleanOutliers['salary_usd'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

jobsDatasetConIQR = jobsDatasetCleanOutliers[(jobsDatasetCleanOutliers['salary_usd'] < lower_bound) | (jobsDatasetCleanOutliers['salary_usd'] > upper_bound)]
print("Outliers según IQR:")
print(len(jobsDatasetConIQR))

DF_jobsDatasetConIQR = jobsDatasetCleanOutliers[(jobsDatasetCleanOutliers['salary_usd'] >= lower_bound) & (jobsDatasetCleanOutliers['salary_usd'] <= upper_bound)]
print("Datos sin outliers:")
print(len(DF_jobsDatasetConIQR))

columna = DF_jobsDatasetConIQR['salary_usd']
plt.figure(figsize=(6, 4))  # Tamaño de la figura
columna.hist(bins=30)  # Puedes cambiar el número de bins según lo que necesites
plt.title('Histograma Amount')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()

sns.boxplot(x=DF_jobsDatasetConIQR['salary_usd'])
plt.show()
```

```python
# Implementacion regresion lineal

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

nombreColumnas = ['job_title', 'experience_level', 'years_experience', 'employment_type', 'education_required', 'remote_ratio', 'company_location', 'company_size','benefits_score', 'salary_usd']
jobsDatesetModeloRegresion = jobsDatasetClean[nombreColumnas].copy()

variableObjetivo = jobsDatesetModeloRegresion['salary_usd']
jobsDatesetModeloRegresion = jobsDatesetModeloRegresion.drop(columns=['salary_usd'])

# Creacion de los conjuntos de datos para el modelo de prediccion
X_train, X_test, y_train, y_test = train_test_split(jobsDatesetModeloRegresion, variableObjetivo, test_size=0.2,
                                                    random_state=42)
model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Evaluación del Modelo:")
print(f"MAE (Error Absoluto Medio): {mae:.2f}")
print(f"MSE (Error Cuadrático Medio): {mse:.2f}")
print(f"R² (Coeficiente de Determinación): {r2:.2f}")

evaluarCasoMediaDataset = pandas.DataFrame({
    'job_title': [9],
    'experience_level': [1],
    'years_experience': [2],
    'employment_type': [1],
    'education_required': [1],
    'remote_ratio': [0.5],
    'company_location': [9],
    'company_size': [1],
    'benefits_score': [7]
}
)

evaluarCasoDos = pandas.DataFrame({
    'job_title': [19],
    'experience_level': [3],
    'years_experience': [4],
    'employment_type': [3],
    'education_required': [3],
    'remote_ratio': [1.0],
    'company_location': [19],
    'company_size': [10],
    'benefits_score': [2]
}
)

evaluarCasoDataset005 = pandas.DataFrame({
    'job_title': [1],
    'experience_level': [0],
    'years_experience': [0.000000],
    'employment_type': [3],
    'education_required': [2],
    'remote_ratio': [1.0],
    'company_location': [6],
    'company_size': [2],
    'benefits_score': [6.6]
}
)

prediccion = model.predict(evaluarCasoMediaDataset)
print("Posible overall-score (media): ", prediccion[0])

prediccionDos = model.predict(evaluarCasoDos)
print("Posible overall-score Caso 2: ", prediccionDos[0])

prediccionTres = model.predict(evaluarCasoDataset005)
print("Posible overall-score Caso 2 (005): ", prediccionTres[0])
```

```python
# EJECUCION MODELO DE AGRUPAMIENTO - KMEANS

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Se eligen las columnas por las cuales se va aplicar el modelo de agrupamiento
nombreColumnas = ['salary_usd', 'years_experience', 'remote_ratio', 'experience_level','education_required','employment_type','company_size','job_description_length']
jobsDatesetKmeans = jobsDatasetClean[nombreColumnas].copy()

# Se aplica un valor de escalado estadar a los datos del dataframe para mejorar la efectividad del KMeans
scaler = StandardScaler()
jobsDatesetKmeansScaler = scaler.fit_transform(jobsDatesetKmeans)

# Creamos el modelo de Kmeans
kmeansUniDataset = KMeans(n_clusters=3, random_state=42)
jobsDatasetClean['Cluster'] = kmeansUniDataset.fit_predict(jobsDatesetKmeansScaler)

jobsDatasetClean.info()

plt.figure(figsize=(10, 6))
plt.scatter(jobsDatasetClean['salary_usd'], jobsDatasetClean['years_experience'], c=jobsDatasetClean['Cluster'], cmap='viridis', s=10)
plt.colorbar(label='Cluster')
plt.title('Clustering de IA Jobs (K-Means)')
plt.xlabel('salary_usd')
plt.ylabel('years_experience')
plt.show()
```
